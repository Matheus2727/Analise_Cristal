```python
import pandas as pd
import os
import numpy as np

caminho = os.path.abspath("dados.csv")
```

```python
df1 = pd.read_csv(caminho)
df1
```

```python
df2 = df1
df2["ID"] = df2.index
df3 = df2
df3["ID_prox"] = df3.index + 1
df3
```

```python
df3r = df3
df3l = df3
df3l["ID"] = df3l["ID_prox"]
df4 = df3r.join(df3l, rsuffix='', lsuffix='_anterior', on="ID")
df4['ID'] = df4['ID'] - 1
df5 = df4[["ID", "leitura", "leitura_anterior"]].dropna()
df5
```

```python
df6 = df5
df6["evento"] = 0
df6["evento"] = np.where(df6['leitura'] >= df6['leitura_anterior'], 0, 1)
df6
```

```python
df7 = df6.loc[df6["evento"] == 1]
df7
```

```python
df8 = df7
df8["var"] = df8["leitura_anterior"] - df8["leitura"]
df8
```

```python
df9 = df8.reset_index()
df9["ID_novo"] = df9.index
df9l = df9
df9l["ID_novo"] = df9l["ID_novo"] + 1
df10 = df9.join(df9l, rsuffix='_proximo', lsuffix='', on="ID_novo")
df11 = df10[["ID", "leitura", "leitura_anterior", "evento", "ID_proximo", "var"]].dropna()
df11
```

```python
df12 = df11
df12["demora"] = df12["ID_proximo"] - df12["ID"]
df12
```

```python
df13 = df12
df13["var_step"] = df12["var"] / df12["demora"]
df14 = df13[["ID", "leitura", "var_step"]]
df14
```

```python
df14.to_csv("trabalhado.csv")
```
